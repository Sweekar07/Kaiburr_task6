```python
import pandas as pd
```

# reading the 70_30 results

```python
df = pd.read_excel("70_30.xlsx")
```

```python
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Setting the position and width for the bars
pos = list(range(len(df['Accuracy'])))
width = 0.3

# Plotting the bars
fig, ax = plt.subplots(figsize=(15, 7))

# Create a bar with Accuracy data
plt.bar(pos, df['Accuracy'], width, alpha=0.5, color='blue', label=df['Algorithm'][0])

# Create a bar with Precision Macro data
plt.bar([p + width for p in pos], df['Precision Macro'], width, alpha=0.5, color='red', label=df['Algorithm'][1])

# Set the y axis label
ax.set_ylabel('Percentage')

# Set the x axis labels
ax.set_xticks([p + 0.5 * width for p in pos])
ax.set_xticklabels(df['Algorithm'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos) - width, max(pos) + width*2)
plt.ylim([0, 100])

# Adding the legend and showing the plot
plt.legend(['Accuracy', 'Precision Macro'], loc='upper left')
plt.grid()

# Add a label to the bar
for bar in ax.containers[0]:
    y_val = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, y_val + 1, f'{y_val:.1f}%', ha='center', va='bottom')

plt.show()
```
